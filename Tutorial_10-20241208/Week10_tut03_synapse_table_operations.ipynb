This tutorial just shows a few useful operations for working with the sample data we provide.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
appos = pd.read_csv("toy_data/10um_appositions.csv")
```

```python
appositions_per_connection = appos.groupby(['@source_node', '@target_node'])['edge_ids'].count()
appositions_per_connection
```

```python
plt.hist(appositions_per_connection, bins=np.arange(1, appositions_per_connection.max()), density=True)
```
